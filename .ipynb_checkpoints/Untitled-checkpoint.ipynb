```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm 
import seaborn as sns 
sns.set()
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.describe(include='all')
```

#cleaning data 

```python
data_drop = data.drop(['id','Unnamed: 0'],axis=1)
data_drop.describe(include='all')
```

```python
data_drop.shape
```

```python
data_drop.info()
```

```python
data_drop.isnull().sum()
```

```python
arrival_delay_median = data_drop["Arrival Delay in Minutes"].median()
data_drop["Arrival Delay in Minutes"] = data_drop["Arrival Delay in Minutes"].fillna( value = arrival_delay_median)
data_no_mv = data_drop
```

```python
data_no_mv.describe(include='all')
```

```python
cols = ["Departure Delay in Minutes", "Arrival Delay in Minutes"]
fig, ax = plt.subplots(2, 1, figsize = (12, 8))
for i, column in enumerate(cols):
    sns.boxplot(data = data_no_mv, x = column,ax = ax[i])
```

```python
sns.distplot(data_no_mv['Departure Delay in Minutes'])
```

```python
sns.distplot(data_no_mv['Arrival Delay in Minutes'])
```

```python
data_no_mv[cols].skew()
```

```python
Q1 = data_no_mv[cols].quantile(0.25)
Q3 = data_no_mv[cols].quantile(0.75)
IQR = Q3 - Q1
minimum=Q1 - 1.5 * IQR
maximum=Q3 + 1.5 * IQR
print(minimum)
print(maximum)
new_data = data_no_mv[data_no_mv['Departure Delay in Minutes'] < maximum[0]]
new_data = data_no_mv[data_no_mv['Arrival Delay in Minutes'] < maximum[1]]
new_data
```
